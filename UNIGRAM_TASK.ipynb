```python
import pandas as pd
df = pd.read_csv("Downloads/archive/UrduNewsHeadlines.csv", encoding='utf-8')
df.head()
```

```python
#Data Pre Processing
df[['Category', 'Headline']] = df[df.columns[0]].str.split('\t', expand=True)
df = df.drop(columns=['Category'])
paragraph = ' '.join(df['Headline'].dropna().tolist())
print(paragraph)
```

```python
#Writing the urdu paragraph in a text file (paragraph form)
with open("urdu_paragraph.txt", "w", encoding="utf-8") as f:
    f.write(paragraph)
#Reading text from the txt file
with open("urdu_paragraph.txt", "r", encoding="utf-8") as f:
    text = f.read()    
```

```python
words = text.split()
vocab_size = len(set(words))
print("Vocabulary Size:", vocab_size)
```

```python
#Reading text from the txt file
with open("urdu_paragraph.txt", "r", encoding="utf-8") as f:
    text = f.read()
#Getting lenght of text characters
print(f"Text length: {len(text)} characters")
print(text)
```

```python
#model training for unigram:
import sentencepiece as spm
spm.SentencePieceTrainer.train(
    input="urdu_paragraph_cleaned.txt",
    model_prefix="urdu_unigram",
    vocab_size=2215,
    model_type="unigram",
    character_coverage=1.0
)
```

```python
sp = spm.SentencePieceProcessor(model_file='urdu_unigram.model')
#Get vocabulary size
vocab_size = sp.get_piece_size()
#Print all tokens
for i in range(vocab_size):
    unitoken = sp.id_to_piece(i)
    print(f"{i}: {unitoken}")
```

```python
#Let's Test the trained Unigram model
unitext = "وزیراعظم نے آج قوم سے خطاب کیا۔"
unitokens = sp.encode(unitext, out_type=str)
print(unitokens)
```
