```python
with open("urdu_paragraph_cleaned.txt", "r", encoding="utf-8") as f:
    text = f.read()

print(f"Text length: {len(text)}")
print(text)
```

```python
#model training for BPE:
import sentencepiece as spm
spm.SentencePieceTrainer.train(
    input="urdu_paragraph_cleaned.txt",    
    model_prefix="urdu_bpe",              
    vocab_size=2215,                      
    model_type="bpe",                     
    character_coverage=1.0          
)
```

```python
sp = spm.SentencePieceProcessor(model_file='urdu_bpe.model')

#Get vocabulary size
vocab_size = sp.get_piece_size()

#Print all tokens
for i in range(vocab_size):
    bpetoken = sp.id_to_piece(i)
    print(f"{i}: {bpetoken}")
```

```python
# Let's test our trained mode
# Test it on an Urdu sentence
bpetext = "وزیراعظم نے آج قوم سے خطاب کیا۔"
bpetokens = sp.encode(bpetext, out_type=str)
print("Tokens:", bpetokens)
```
